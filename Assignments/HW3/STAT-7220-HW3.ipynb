7. Fit a full factorial ANOVA model to the data. Assess the normality assumption using both a visual method and a testing method. Do we have evidence to support the assumption of normality? Why or why not?

**A full factorial ANOVA model was fit to the data, including the main effects of sole material, insole thickness, and breathability, as well as all interaction terms.**

**The normality assumption was assessed using both a Q-Q plot and the Shapiro-Wilk test. The Q-Q plot showed that the residuals were approximately along the reference line, suggesting that the residuals are reasonably normal.**

**The Shapiro-Wilk test produced a p-value of 0.390. Since this p-value is greater than 0.05, we fail to reject the null hypothesis that the residuals are normally distributed.Therefore, there is no evidence against the normality assumption, and the assumption appears to be satisfied..**

```python

import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats



mod = ols('comfort ~ sole_material*insole_thickness*breathability', data=shoe).fit()

print(mod.summary())

## Test Normality Assumption ##

residuals = mod.resid

## Generate QQ plot using seaborn ##

sm.qqplot(residuals, line='s')
plt.title("QQ Plot of Model Residuals")
plt.style.use('classic')
plt.show()

## S-W Test ##

sw = stats.shapiro(residuals)
sw_statistic = sw[0]
sw_p_value = sw[1]

print(f"S-W Statistic: {sw_statistic}")
print(f"S-W p-value: {sw_p_value}")

if sw_p_value < 0.05:
    print("Reject the null hypothesis: Residuals are not normally distributed.")
else:
    print("Fail to reject the null hypothesis: Residuals are normally distributed.")
```


8.  Assess the constant variance assumption using both a visual method and a testing method. Do we have evidence to support the assumption of constant variance? Why or why not?

**The constant variance assumption was assessed using a residuals versus fitted values plot and the Breusch-Pagan test.The residual plot shows that the points are randomly scattered around zero with no clear pattern or funnel shape, suggesting that the variance is approximately constant.The Breusch-Pagan test resulted in a p-value of 0.0818. Since this p-value is greater than 0.05, we fail to reject the null hypothesis of constant variance. Therefore, there is no evidence to suggest that the variance is not constant, and the assumption of constant variance appears to be satisfied.**

```python

    ## Constant Variance ##

## Obtain Standardized Residuals ##
standardized_residuals = mod.get_influence().resid_studentized_internal

## Get the fitted values ##
fitted_values = mod.fittedvalues

## Create the plot ##
plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(mod.resid, mod.model.exog)

f_statistic = bp_test[2]
f_p_value = bp_test[3]

print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")
if f_p_value < 0.05:
  print("Reject the null hypothesis: Residuals are not homoscedastic.")
else:
  print("Fail to reject the null hypothesis: Residuals are homoscedastic.")
```
