6.  Perform appropriate exploratory analysis, including data visualizations. Do the results of these analyses support the null or alternative hypotheses more strongly?

```python

## Import pandas library ##
import pandas as pd
import numpy as np

## Import Shoe Comfort Experiment.xlsx File ##
shoe = pd.read_excel("../Assignments/HW3/Shoe Comfort Experiment.xlsx")

## Data Integrity Check ##
print(shoe.info())

## Calculate Means & SDs of Sole Material ##
sole_stats = shoe.groupby("sole_material")["comfort"].agg(['mean','std'])
print(sole_stats)

## Means and Standard Deviations for Insole Thickness ##
insole_stats = shoe.groupby("insole_thickness")["comfort"].agg(['mean','std'])
print(insole_stats)

## Means and Standard Deviations for Breathability ##
breath_stats = shoe.groupby("breathability")["comfort"].agg(['mean','std'])
print(breath_stats)

## Interaction Effect Summary Stats ##
int_stats = shoe.groupby(['sole_material','insole_thickness'])['comfort'].agg(['mean','std'])
print(int_stats)

## Import seaborn and matplotlib ##
import seaborn as sns
import matplotlib.pyplot as plt

#################################
## Create the boxplot using seaborn ##
sns.set_style("whitegrid")

sns.boxplot(
    data=shoe,
    x='sole_material',
    y='comfort',
    hue='insole_thickness'
)

## Calculate Means Per Sole Material ##
mean_points = shoe.groupby('sole_material')['comfort'].mean().reset_index()

## Overlay Mean Points ##
sns.scatterplot(data=mean_points,x="sole_material",y="comfort",
                marker='X',s=100,color='black',label='Mean',
                zorder=3)

## Add labels and titles ##
plt.title('Boxplot of Comfort by Sole Material with Insole Thickness Fill and Mean Overlay')
plt.ylabel('Comfort Rating')
plt.xlabel('Sole Material')
plt.legend()
plt.tight_layout()
plt.show()

###############################################
## First, let's grab the interaction means ##
int_means = shoe.groupby(['sole_material','insole_thickness'])['comfort'].mean().reset_index()

## Next, Create Interaction Plot with Seaborn ##
sns.set_style('whitegrid')

sns.lineplot(data=int_means,
             x='sole_material',
             y='comfort',
             hue='insole_thickness',
             marker='o',
             markersize=10,
             style='insole_thickness')

## Add Titles and Labels ##
plt.title('Mean Comfort by Sole Material and Insole Thickness')
plt.ylabel('Comfort Rating')
plt.xlabel('Sole Material')
plt.show()

#########################
## Step 1: Create a figure and a single axes object to plot on. ##
fig, ax = plt.subplots(figsize=(12, 8))

## Step 2: Plot the boxplot first, on the raw data. ##
sns.boxplot(data=shoe, x='sole_material', y='comfort', hue='insole_thickness',
            palette='viridis', dodge=True, ax=ax, zorder=2)

## Step 3: Overlay the interaction plot (line plot of the means). ##
sns.lineplot(data=shoe, x='sole_material', y='comfort', hue='insole_thickness',
             marker='o', markersize=10, ax=ax, zorder=3,
             estimator='mean', err_style=None, linestyle='--',
             palette='dark')

## Step 4: Clean up the plot: set titles, labels, and theme. ##
ax.set_title('Interaction Plot of Mean Comfort Overlaid on Boxplot', fontsize=16)
ax.set_xlabel('Sole Material', fontsize=12)
ax.set_ylabel('Comfort Rating', fontsize=12)

## Adjust legend to avoid duplicate labels ##
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2], title='Insole Thickness')

## Apply a classic theme by removing the top and right spines. ##
sns.despine()
plt.tight_layout()

## Show Plot ##
plt.show()
```

7. Fit a full factorial ANOVA model to the data. Assess the normality assumption using both a visual method and a testing method. Do we have evidence to support the assumption of normality? Why or why not?

**A full factorial ANOVA model was fit to the data, including the main effects of sole material, insole thickness, and breathability, as well as all interaction terms.**

**The normality assumption was assessed using both a Q-Q plot and the Shapiro-Wilk test. The Q-Q plot showed that the residuals were approximately along the reference line, suggesting that the residuals are reasonably normal.**

**The Shapiro-Wilk test produced a p-value of 0.390. Since this p-value is greater than 0.05, we fail to reject the null hypothesis that the residuals are normally distributed.Therefore, there is no evidence against the normality assumption, and the assumption appears to be satisfied..**

```python
## Import relevant statsmodels and scipy functions ##
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats

## Fit the ANOVA Model ##

mod = ols('comfort ~ sole_material*insole_thickness*breathability', data=shoe).fit()

## Print Model Summary ##
print(mod.summary())

## Test Normality Assumption ##

## QQ-Plot ##

## Extract Model Residuals ##

residuals = mod.resid

## Generate QQ plot using seaborn ##

sm.qqplot(residuals, line='s')
plt.title("QQ Plot of Model Residuals")
plt.style.use('classic')
plt.show()

## S-W Test ##

sw = stats.shapiro(residuals)
sw_statistic = sw[0]
sw_p_value = sw[1]

print(f"S-W Statistic: {sw_statistic}")
print(f"S-W p-value: {sw_p_value}")

if sw_p_value < 0.05:
    print("Reject the null hypothesis: Residuals are not normally distributed.")
else:
    print("Fail to reject the null hypothesis: Residuals are normally distributed.")
```


8.  Assess the constant variance assumption using both a visual method and a testing method. Do we have evidence to support the assumption of constant variance? Why or why not?

**The constant variance assumption was assessed using a residuals versus fitted values plot and the Breusch-Pagan test.The residual plot shows that the points are randomly scattered around zero with no clear pattern or funnel shape, suggesting that the variance is approximately constant.The Breusch-Pagan test resulted in a p-value of 0.0818. Since this p-value is greater than 0.05, we fail to reject the null hypothesis of constant variance. Therefore, there is no evidence to suggest that the variance is not constant, and the assumption of constant variance appears to be satisfied.**

```python

    ## Constant Variance ##

## Obtain Standardized Residuals ##
standardized_residuals = mod.get_influence().resid_studentized_internal

## Get the fitted values ##
fitted_values = mod.fittedvalues

## Create the plot ##
plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(mod.resid, mod.model.exog)

f_statistic = bp_test[2]
f_p_value = bp_test[3]

print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")
```


9.  Report the F-statistics and their associated p-values from the factorial ANOVA model. What conclusions can we draw about our main effects? Our two-way interactions? Our three-way interaction?

**The F-statistics and associated p-values from the factorial ANOVA model were examined for each main effect and interaction.**

**For the main effects, sole material had an F-statistic of 5.813 with a p-value of 0.028, indicating that it is statistically significant. In contrast, insole thickness (F = 0.691, p = 0.418) and breathability (F = 0.085, p = 0.775) were not statistically significant. This suggests that sole material has a significant effect on comfort ratings, while insole thickness and breathability do not.**

**For the two-way interactions, none were statistically significant. Specifically, the interaction between sole material and insole thickness (F = 0.312, p = 0.584), sole material and breathability (F ≈ 0, p = 1.000), and insole thickness and breathability (F = 0.999, p = 0.332) were all not significant. This indicates that there is no evidence that the effect of one factor depends on the level of another factor.**

**For the three-way interaction, the interaction between sole material, insole thickness, and breathability (F = 0.691, p = 0.418) was also not statistically significant. This suggests that there is no combined interaction effect among the three factors.**

**Overall, the results indicate that sole material is the only significant factor affecting comfort, and there is no evidence of interaction effects.**

```python
shoe_tab = sm.stats.anova_lm(mod, typ=3)

shoe_tab["Significance"] = np.where(shoe_tab['PR(>F)'] < 0.05,
                                    'Significant',
                                    'Not Significant')

print(shoe_tab.loc[['sole_material',
                    'insole_thickness',
                    'breathability',
                    'sole_material:insole_thickness',
                    'sole_material:breathability',
                    'insole_thickness:breathability',
                    'sole_material:insole_thickness:breathability']])
```


10. Calculate and interpret the partial $\eta^2$ values for each of the main effects and interactions. Which effects and/or interactions are most important? Why?

**Partial η² values were calculated to assess the relative importance of each main effect and interaction in explaining variability in comfort ratings.**

**The results show that sole material has the largest partial η² value (η² = 0.266), indicating that it explains the greatest proportion of variability in comfort ratings. In comparison, insole thickness (η² = 0.041) and breathability (η² = 0.005) have much smaller values, suggesting they contribute less to the variation in comfort.**

**Among the interaction effects, all partial η² values are relatively small. The largest interaction effect is between insole thickness and breathability (η² = 0.059), but it is still much smaller than the effect of sole material. The remaining interaction effects are minimal and do not meaningfully contribute to explaining variability in the response.**

**Overall, sole material is the most important factor, as it accounts for the largest share of variability in comfort ratings, while the other factors and interactions have relatively minor effects.**


```python
## Calculate Partial Eta Squared ##

# Extract residual sum of squares
ss_res = shoe_tab.loc['Residual', 'sum_sq']

# Compute partial eta squared for each effect
eta_sq = shoe_tab['sum_sq'] / (shoe_tab['sum_sq'] + ss_res)

# Add to table
shoe_tab['Partial_Eta_Sq'] = eta_sq

# Print relevant rows
print(shoe_tab.loc[['sole_material',
                    'insole_thickness',
                    'breathability',
                    'sole_material:insole_thickness',
                    'sole_material:breathability',
                    'insole_thickness:breathability',
                    'sole_material:insole_thickness:breathability']])
```
